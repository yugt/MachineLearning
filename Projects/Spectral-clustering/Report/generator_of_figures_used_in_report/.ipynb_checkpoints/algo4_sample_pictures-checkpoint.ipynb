# Spectral Clustering Based on Local PCA
The main algorithms and functions are describe in "spectral_clustering_lpca.py"

### parameters
- $r$ : Shoulde be chosen to be a little larger than the line/surface thickness 
- $eps$ : Close to $r$
- $eta$ : Close to 1 or less than 1. Unlike $r$ and $eps$, it doesn't depend on the size of the picture. When $eta$ is small (close to 0), "direction" of the groups will have a strong effect.
- $d$ : Dimension of the hyperplane of the group. If we want to separete written figures, set $d$ = 1 because it is just a 1-dimensional line. 
- $K$ : The number of groups which we want to separete the original data into.

```python
import spectral_clustering_lpca as sc
# %matplotlib inline
```

## Example 1
The original "720" figure Guangting prepared.

Looks nice. But the intersection part is totally included in "0" part.

```python
sc.plot_spectral_clustering(picture_name ='connected-digits.png',
                         color_threshold = 10,
                         r = 25.0,
                         eps = 25.0,
                         eta = 1.0,
                         d = 1,
                         K = 3)
```

## Example 2
This is an example of the small $eta$. The data is separeted based on their "local direction".
- Green: Horizontal direction
- Blue : Upper right diagonal direction
- Red  : Upper left diagonal direction

```python
sc.plot_spectral_clustering(picture_name ='connected-digits.png',
                         color_threshold = 10,
                         r = 25.0,
                         eps = 25.0,
                         eta = 0.1,
                         d = 1,
                         K = 3)
```
