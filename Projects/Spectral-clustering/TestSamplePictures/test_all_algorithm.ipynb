parameters

- $algorithm$ : 2 stands for comparing convariance. 3 stands for comparing projection. 4 stands for spectral clustering based on local PCA. The default is algorithm 4.
- $r$ : Shoulde be chosen to be a little larger than the line/surface thickness
- $eps$ : Close to $r$
- $eta$ : Close to 1 or less than 1. Unlike $r$ and $eps$, it doesn't depend on the size of the picture. When $eta$ is small (close to 0), "direction" of the groups will have a strong effect.
- $d$ : Dimension of the hyperplane of the group. If we want to separete written figures, set $d$ = 1 because it is just a 1-dimensional line. This parameter is only used in algorithm 4.
- $K$ : The number of groups which we want to separete the original data into. This parameter is only used in algorithm 4.
